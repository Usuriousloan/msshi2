```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# ===  1:   ===
np.random.seed(42)
X = np.sort(np.random.uniform(-10, 10, 1000)).reshape(-1, 1)
y = np.sin(X).ravel() + np.random.normal(0, 0.1, 1000)

plt.figure(figsize=(8, 6))
plt.scatter(X, y, s=15, alpha=0.6)
plt.title(' ')
plt.xlabel('X')
plt.ylabel('y')
plt.grid(True)
plt.show()

# ===  2:  ===
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# ===  3:   Train/Test ===
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

plt.figure(figsize=(8, 6))
plt.scatter(X_train, y_train, s=15, alpha=0.6, label='train')
plt.scatter(X_test, y_test, s=15, alpha=0.6, label='test')
plt.title(' ')
plt.xlabel('X ()')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# ===  4:     K (Train vs Test MSE) ===
k_values = range(1, 21) 
mse_train_list = []
mse_test_list = []

for k in k_values:
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train, y_train)
    
    y_pred_train = knn.predict(X_train)
    mse_train_list.append(mean_squared_error(y_train, y_pred_train))
    
    y_pred_test = knn.predict(X_test)
    mse_test_list.append(mean_squared_error(y_test, y_pred_test))

plt.figure(figsize=(8, 6))
plt.plot(k_values, mse_train_list, label='Train MSE')
plt.plot(k_values, mse_test_list, label='Test MSE')
plt.title('    ')
plt.xlabel('k')
plt.ylabel('MSE')
plt.legend()
plt.grid(True)
plt.show()

# ===  5:   K ===
best_index = np.argmin(mse_test_list)
best_k = k_values[best_index]
min_mse = mse_test_list[best_index]

plt.figure(figsize=(8, 6))
plt.plot(k_values, mse_train_list, label='Train MSE')
plt.plot(k_values, mse_test_list, label='Test MSE')
plt.axvline(best_k, color='red', linestyle='--', label=f'Best k={best_k}')
plt.title('MSE vs k ( )')
plt.xlabel('k')
plt.ylabel('MSE')
plt.legend()
plt.grid(True)
plt.show()

print(f" K: {best_k}")
print(f" MSE: {min_mse:.5f}")

# ===  6:    ===
knn_best = KNeighborsRegressor(n_neighbors=best_k)
knn_best.fit(X_train, y_train)

X_grid = np.linspace(0, 1, 500).reshape(-1, 1)
y_pred_grid = knn_best.predict(X_grid)

plt.figure(figsize=(8, 6))
plt.scatter(X_train, y_train, s=15, alpha=0.6, label='train', color='tab:blue')
plt.scatter(X_test, y_test, s=15, alpha=0.6, label='test', color='tab:orange')
plt.plot(X_grid, y_pred_grid, color='red', linewidth=2, label=f'Prediction k={best_k}')
plt.title(' KNN ')
plt.xlabel('X ()')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
